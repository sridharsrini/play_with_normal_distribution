```python
import numpy as np
```

# card experiment 1
draw a card, if black 0 and if red 1

```python
# generate a random integer number between low and high
num = np.random.randint(low=0,high=1+1,size=10)
print(num)
```

```python
# using the above construct, let us simulate a card draw experiment mapping black to 0 and 1 to red
# conduct this experiment a large number of times and count the # of 0s and 1s
# create an array with various values of n
n = np.array([10,100,1000,10000,100000])
for i in n:
    num = np.random.randint(low=0,high=1+1,size=i)
    print("from ",i," trials ", "count of 1s is :", sum(num)," and 0s is: ", i - sum(num))
```

We can see that over large # of trials the counts of 0s and 1s tends to become equal. Hence their probabilities are 0.5.

# card experiment 2
place bet on a suite of cards and then draw. 100 if the matching suite is drawn else -100.

```python
# set up the experiment using random number generators
# assume the bet is placed on a suite that we link to number 1. The other three suites are linked to 2,3,4
# this statement generates a number between 1 and 4 inclusive
num = np.random.randint(low=1,high=4+1,size=10)
print(num) # this is the array generated
print(num==1) # this marks each element as true if it is equal to 1
print(sum(num==1)) # this totals up the true elements. Python assigns true=1 and false=0
```

```python
# place a bet on a number between 1 and 4 corresponding to [1] diamonds, [2] hearts, [3] clubs, [4] spades
bet = np.random.randint(low=1,high=4+1,size=1)
print("bet was placed on number ", bet)
# let us repeat this many times and see the results
n = np.array([10,100,1000,10000,100000])
for i in n:
    num = np.random.randint(low=1,high=4+1,size=i)
    print("from ",i," trials ", "count of 1s is :", sum(num==bet)," and 2-4s is: ", i - sum(num==bet))
```

 we can see that as the no of trials increases the proportion is 25% and 75%, hence the corresponding probabilities
